```python
# Graph 구조를 팀원과 팀원 친구의 이름으로 바꿔서 실습
from collections import deque

new_graph = {
    '고용민': ['김진서', '양지원', '강예근'],
    '김진서': ['박진영', '이유라'],
    '양지원': ['김영휘'],
    '김영휘': ['고용민', '강수진'],
    '강예근': ['이유라', '박민혁'],
    '이유라': ['이서경'],
    '이서경': ['박민혁'],
    '박민혁': ['강예근'],
    '박진영': ['김예진'],
    '김예진': ['양지원'],
    '강수진': ['박진영']
}

# 2. DFS 함수
def dfs(graph, v, visited):
  visited.add(v)
  print(v+"->", end=' ')

  # v(이름)가 graph 딕셔너리의 key로 존재할 때만 실행
  if v in graph:
    for neighbor in graph[v]:
      if neighbor not in visited:
        dfs(graph, neighbor, visited)

# 3. DFS 실행
print("DFS 탐색순서")
dfs(new_graph, '고용민', set())
print() # 줄바꿈
```

```python
# 감염 시뮬레이션으로 확장 (TC - 1)

from collections import deque

def find_infections_bfs(graph, start_node, max_level=5):


    # 1. 큐(Queue)와 방문(Visited) 기록, 결과 리스트 초기화
    # 큐에는 (사람, 현재 레벨) 튜플을 저장합니다.
    queue = deque([(start_node, 0)])

    # visited는 방문 여부와 함께 *최단* 레벨을 저장합니다.
    # 시작 노드는 0레벨로 방문 처리
    visited = {start_node: 0}

    # 결과 리스트 (감염자 자신은 제외, 조건 2)
    infected_list = []

    # 2. 큐가 빌 때까지 BFS 탐색 시작
    while queue:
        current_person, current_level = queue.popleft()

        # 3. 최대 레벨(5단계)에 도달하면 더 이상 탐색하지 않음
        # (현재 노드의 *다음* 레벨이 max_level을 초과하면 안 됨)
        if current_level >= max_level:
            continue

        # 4. 현재 사람의 모든 접촉자(이웃) 확인
        for neighbor in graph.get(current_person, []):

            # 5. 아직 방문하지 않은 사람인 경우
            if neighbor not in visited:

                new_level = current_level + 1

                # 방문 처리 (해당 노드의 레벨 기록)
                visited[neighbor] = new_level

                # 큐에 추가하여 다음 탐색 대상으로 등록
                queue.append((neighbor, new_level))

                # 결과 리스트에 (감염자, 레벨) 추가 (조건 2, 3 충족)
                infected_list.append((neighbor, new_level))

    return infected_list

# --- 제공된 그래프 데이터 (graph_20) ---
graph_10 = {
    '고용민': ['김진서', '양지원', '강예근'],
    '김진서': ['박진영', '이유라'],
    '양지원': ['김영휘'],
    '김영휘': ['고용민', '강수진'],
    '강예근': ['이유라', '박민혁'],
    '이유라': ['이서경'],
    '이서경': ['박민혁'],
    '박민혁': ['강예근'],
    '박진영': ['김예진'],
    '김예진': ['양지원'],
    '강수진': ['박진영']
}



# --- 시뮬레이션 실행 ---
start_person = '고용민'
max_depth_limit = 2 #2단계 이내 감염경로 추적

infections = find_infections_bfs(graph_10, start_person, max_level=max_depth_limit)

# --- 결과 출력 (형식 자유롭게, 레벨 표시) ---
print(f"--- 💻 '{start_person}' 기준 {max_depth_limit}단계 이내 감염 경로 추적 (BFS) ---")
print(f"최초 감염자: {start_person}\n")



# 레벨별로 결과를 그룹화하여 출력
results_by_level = {}
for person, level in infections:
    if level not in results_by_level:
        results_by_level[level] = []
    results_by_level[level].append(person)

# 정렬된 레벨 순서로 출력
if not results_by_level:
    print("   감염자가 없습니다.")
else:
    for level in sorted(results_by_level.keys()):
        persons = ", ".join(results_by_level[level])
        print(f"🔵 {level}단계 감염 (거리 {level}):")
        print(f"   {persons}")
```
