```python
#1. 함수 저으이와 시각화
import matplotlib.pyplot as plt
import numpy as np

def f(x): #함수
  return 2*x+1

x = np.linspace(-10,10,100)
y = f(x)

plt.plot(x,y,label='f(x) = 2x+1')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Example of function graph')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#2. 단사 함수 여부 확인
def f(x): #함수
  return 2*x+1

def is_injective(f,domain):
  image = set()
  for x in domain:
    fx = f(x)
    if fx in image:
      return False
    image.add(fx)
  return True

domain = range(-10,11) #정의역
print("Is it injective func?", is_injective(f,domain))
```

```python
#3. 역함수 구하기
import matplotlib.pyplot as plt
import numpy as np
def inverse_f(y):
  return (y-1)/2

y_values = np.linspace(-10,10,100)
x_values = inverse_f(y_values)

plt.plot(y_values,x_values,label = 'Inverse Func f^{-1}(y) = (y-1)/2',color = 'red')
plt.xlabel('y')
plt.ylabel('f^{-1}(y)')
plt.title('Inverse func.graph')
plt.legend()
plt.grid(True)
plt.show()
```
